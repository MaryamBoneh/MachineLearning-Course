```python
import numpy as np
import pandas as pd
import random
import sklearn
import matplotlib.pyplot as plt
from scipy import stats
from numpy.linalg import inv
from sklearn.neighbors import KNeighborsClassifier
```

```python
N = 100

X = []
Y = []

for i in range(N):
  X.append(np.random.randint(60, 200, dtype=int))
  temp = random.uniform(0.25, 0.5) * X[i]
  Y.append(temp)

plt.scatter(X, Y)
```

```python
# m = (X.XT)^-1 *yXT
X = np.array(X)
Y = np.array(Y)

X = X.reshape(100, 1)
Y = Y.reshape(100, 1)

print(X.shape, Y.shape)
```

```python
m1 = random.uniform(0.1, 0.7)
m2 = random.uniform(0.1, 0.7)

Y_pred1 = X * m1
Y_pred2 = X * m2

plt.scatter(X, Y)

plt.plot(X, Y_pred1, c= 'red')
plt.plot(X, Y_pred2, c= 'red')
```

```python
m = np.matmul(inv(np.matmul(X.T, X)), np.matmul(X.T, Y))

Y_pred = X * m

plt.scatter(X, Y)
plt.plot(X, Y_pred, c= 'green')
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(X, Y)

plt.plot(X, Y, 'o', label='original data')
plt.plot(X, intercept + slope*X, 'r', label='fitted line')
plt.legend()
plt.show()
```
